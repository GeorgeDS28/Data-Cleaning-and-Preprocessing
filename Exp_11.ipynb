```python
import pandas as pd
df = pd.read_csv("titanic.csv")
```

```python
#1
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
#2
df['Fare'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())
df.head()
```

```python
#3
df['is_minor'] = df['Age'].apply(lambda x: 1 if x < 18 else 0)

df.head()
```

```python
import pandas as pd
df = pd.read_csv("Iris.csv")
```

```python
#4
df = df.fillna(df.median(numeric_only=True))
df.head()
```

```python
#5
df = pd.get_dummies(df, columns=['Species'])
df.head()
```

```python
#6
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['SepalLengthCm'] = scaler.fit_transform(df[['SepalLengthCm']])
df.head()
```

```python
import pandas as pd
df = pd.read_csv("boston.csv")
```

```python
df.head()
```

```python
#7
df = df.dropna(subset=['chas'])
df.head()
```

```python
#8
Q1 = df['tax'].quantile(0.25)
Q3 = df['tax'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['tax'] >= lower) & (df['tax'] <= upper)]

```

```python
#9
df['dis'].fillna(df['dis'].median(), inplace=True)
df.head()
```

```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
#10

df = df.drop_duplicates()
df.head()
```

```python
#12
from scipy import stats

df = df[(abs(stats.zscore(df['pH'])) < 3)]
df.head()
```
